```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv("student dropout.csv")
```

```python
df
```

```python
df.columns
```

```python
df.isnull().sum()
```

```python
#Finding Non_numerical Columns
non_num = df.select_dtypes(exclude=[np.number])
print(non_num)
```

```python
cat_cols = ['School', 'Gender', 'Address', 'Family_Size', 'Parental_Status', 'Mother_Job', 'Father_Job', 'Reason_for_Choosing_School', 'Guardian', 'School_Support', 'Family_Support', 'Extra_Paid_Class', 'Extra_Curricular_Activities', 'Attended_Nursery', 'Wants_Higher_Education', 'Internet_Access', 'In_Relationship', 'Dropped_Out']
```

```python
new_df = pd.get_dummies(df[cat_cols], drop_first=False).astype(int)
```

```python
new_df
```

```python
df
```

```python
df.drop(df[cat_cols],axis=1,inplace=True)
```

```python
new_data = pd.concat([df,new_df],axis=True)
```

```python
new_data

```

```python
new_data.columns
```

```python
aa= new_data['Dropped_Out']
```

```python
aa
```

```python
#Modelling
```

```python
x=new_data.drop('Dropped_Out',axis=1)
y=new_data['Dropped_Out']

```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
x=sc.fit_transform(x)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
print(x_train.shape, y_train.shape)
```

```python
from sklearn.linear_model import LogisticRegression
lr=LogisticRegression()
lr.fit(x_train,y_train)
```

```python
y_pred=lr.predict(x_test)
```

```python
y_pred
```

```python
lr.score(x_train,y_train)
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
```

```python
confusion_matrix(y_test,y_pred)
```

```python
accuracy_score(y_test,y_pred)
```

```python
report=classification_report(y_test,y_pred)
```
